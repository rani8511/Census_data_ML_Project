## 1. Introduction to Hypothesis Testing

### What is Hypothesis Testing?

Hypothesis testing in machine learning is a statistical method used to validate assumptions, compare model performance, or assess feature significance using sample data.

In simple words, hypothesis testing compares two opposite ideas about a group of people or things and uses data from a small part of that group (a sample) to decide which idea is more likely true. We collect and study the sample data to check if the claim is correct.

It helps determine if observed differences (e.g., higher accuracy from Model A vs. Model B) are statistically significant or just random occurrences.



### Purpose
- It helps in comparing models (e.g., Model A vs. Model B)
- Evaluating feature importance
- Verifying if a model generalizes well

### Components

#### Null Hypothesis (H₀ or H₀)
It is a foundational statistical statement assuming no relationship, difference, or effect exists between variables, positing that any observed data differences are due to chance.  
It serves as a baseline for testing and is either rejected or supported based on p-value analysis against an alternative hypothesis.

It assumes no effect or no difference (e.g., model accuracy is the same).

#### Alternative Hypothesis (Hₐ or H₁)
It is a statement that contradicts the null hypothesis, proposing that there is a significant effect, relationship, or difference in the population.

It represents the research claim the analyst aims to support and is tested by collecting data to determine if the null hypothesis can be rejected.

It assumes a significant effect exists (e.g., one model is superior).

---

## 2. Common Hypothesis Tests by Data Type & Goal



### Numerical Data (Means & Distributions)
- **One-Sample T-test**: Compares a sample mean to a known population mean.
- **Two-Sample T-test (Welch’s)**: Compares means of two independent groups.
- **Paired T-test**: Compares means from the same group at different times (e.g., pre/post).
- **ANOVA (Analysis of Variance)**: Tests differences in means across three or more groups.
- **Mann-Whitney U Test**: Non-parametric version of the two-sample t-test for non-normal data.
- **Kruskal-Wallis Test**: Non-parametric version of ANOVA for three or more groups.

### Categorical Data (Frequencies & Proportions)
- **Chi-Square Test of Independence**: Evaluates if two categorical variables are related.
- **Chi-Square Goodness-of-Fit Test**: Tests if sample data matches an expected distribution.
- **Fisher’s Exact Test**: Used instead of Chi-square for small sample sizes.

### Correlation and Modeling
- **Pearson Correlation Test**: Tests the linear relationship between two continuous variables.
- **Spearman Rank Correlation**: Tests monotonic relationships (non-parametric).
- **Regression Slope Test**: Tests if the slope of a regression line is significantly different from zero.

---

## 3. Key Terms of Hypothesis Testing

To understand hypothesis testing, the following key terms are important:

- **Significance Level (α)**:  
  How sure we want to be before saying the claim is false. Usually set to 0.05 (5%).

- **p-value**:  
  The probability of seeing the observed data if the null hypothesis is true.

### Interpretation
- **p ≤ 0.05**: Statistically significant → Reject the null hypothesis.
- **p > 0.05**: Not statistically significant → Fail to reject the null hypothesis.

- **Test Statistic**:  
  A numerical value that helps decide whether the data supports or rejects the hypothesis.

- **Critical Value**:  
  The cutoff point used to compare against the test statistic.

- **Degrees of Freedom**:  
  A value based on sample size that helps determine the critical value.

---

## 4. Steps of Hypothesis Testing & Hypothesis Generation

### Steps of Hypothesis Testing
1. **State the Hypotheses**  
   Define the Null Hypothesis (H₀: no effect/difference) and Alternative Hypothesis (Hₐ: expected effect/relationship).

2. **Set Decision Criteria**  
   Choose a significance level (α, usually 0.05).

3. **Select & Calculate Test Statistic**  
   Choose the appropriate test based on data type and calculate the statistic (e.g., p-value).

4. **Make a Decision**  
   If p ≤ α, reject H₀; otherwise, fail to reject H₀.

5. **Conclusion**  
   Interpret the result in the context of the original research question.

### How to Generate Hypotheses from a Dataset
- Identify the **target variable** (what we are predicting).
- Identify **key features** (demographic, behavioral, clinical, etc.).
- Analyze whether each feature influences or is associated with the target variable.

# TIER 1 HYPOTHESIS

### Prerequisite Checks for Hypothesis Testing


```python
# 1. Importing Libraries
import pandas as pd
from scipy.stats import chi2_contingency, ttest_ind

# 2. Loading Dataset
adult_df = pd.read_csv(
    "adult.data",
    header=None,
    names=[
        "age", "workclass", "fnlwgt", "education", "education_num",
        "marital_status", "occupation", "relationship", "race", "sex",
        "capital_gain", "capital_loss", "hours_per_week",
        "native_country", "income"
    ],
    na_values="?",
    skipinitialspace=True
)

print("Dataset loaded. Shape:", adult_df.shape)

# 3. Verifying Required Columns
required_cols = ["education", "workclass", "income", "hours_per_week"]
missing_cols = [col for col in required_cols if col not in adult_df.columns]

if missing_cols:
    raise ValueError(f"Missing required columns: {missing_cols}")
else:
    print("All required columns are present.")

# 4. Checking for Missing Values
print("\nMissing values check:")
print(adult_df[required_cols].isnull().sum())

# Removing rows with missing values in required columns
adult_df = adult_df.dropna(subset=required_cols)

# 5. Verifying Data Types
print("\nData types:")
print(adult_df[required_cols].dtypes)
```

## Hypothesis 1: Education vs Income

- **Null Hypothesis (H₀):** There is no association between education level and income.
- **Alternative Hypothesis (H₁):** There is an association between education level and income.
- **Test Used:** Chi-square test of independence, since both education and income are categorical variables.


```python
contingency1 = pd.crosstab(adult_df['education'], adult_df['income'])

chi2, p, dof, expected = chi2_contingency(contingency1)

print("Chi-square statistic:", chi2)
print("p-value:", p)

if p < 0.05:
    print("Decision: Reject the null hypothesis (H₀)")
else:
    print("Decision: Fail to reject the null hypothesis (H₀)")
```

Since p value is less than level of significance (0.05) ,i.e. the probability of this happening is extremely unlikely under null hypothesis is true ,Therefore We reject the null hypothesis . Education has an effect on income.

## Hypothesis 2: Workclass vs Income

- **Null Hypothesis (H₀):** There is no association between workclass and income.
- **Alternative Hypothesis (H₁):** There is an association between workclass and income.
- **Test Used:** Chi-square test of independence, since both workclass and income are categorical variables.


```python
contingency2 = pd.crosstab(adult_df['workclass'], adult_df['income'])

chi2, p, dof, expected = chi2_contingency(contingency2)
print("p-value:", p)
print("Chi-square statistic:", chi2)

if p < 0.05:
    print("Decision: Reject the null hypothesis (H₀)")
else:
    print("Decision: Fail to reject the null hypothesis (H₀)")
```

Since p value is less than level of significance (0.05) ,i.e. the probability of this happening is extremely unlikely under null hypothesis is true ,Therefore We reject the null hypothesis . Workclass has an effect on income.

## Hypothesis 3: Hours per Week vs Income

- **Null Hypothesis (H₀):** The mean number of working hours per week is the same across income groups.
- **Alternative Hypothesis (H₁):** The mean number of working hours per week differs across income groups.
- **Test Used:** Independent t-test (Welch’s t-test), since hours per week is a numeric variable and income has two independent groups.


```python
high_income_work_hours = adult_df[adult_df['income'] == '>50K']['hours_per_week']
low_income_work_hours = adult_df[adult_df['income'] == '<=50K']['hours_per_week']

t_stat, p = ttest_ind(
    high_income_work_hours,
    low_income_work_hours,
    equal_var=False
)

print("t-statistic:", t_stat)
print("p-value:", p)

if p < 0.05:
    print("Decision: Reject the null hypothesis (H₀)")
else:
    print("Decision: Fail to reject the null hypothesis (H₀)")
```

Since p value is less than level of significance (0.05) ,i.e. the probability of this happening is extremely unlikely under null hypothesis is true ,Therefore We reject the null hypothesis . Hours-per-week different for different income groups.
