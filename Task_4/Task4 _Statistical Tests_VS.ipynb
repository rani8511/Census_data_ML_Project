## Phase 4 — Statistical Tests & Hypothesis Validation

In this phase, we validate the hypotheses generated in Phase 3 using appropriate statistical tests.  
We also explain hypothesis testing concepts, p-values, and Type I / Type II errors.

---

### Hypothesis Testing Overview

- **Null Hypothesis (H₀):** Assumes no relationship or no difference exists.
- **Alternative Hypothesis (H₁):** Assumes a relationship or difference exists.

### p-value
- The probability of observing the data assuming the null hypothesis is true.
- **Decision Rule:**
  - If p-value ≤ 0.05 → Reject H₀
  - If p-value > 0.05 → Fail to reject H₀

### Type I Error (False Positive)
- Rejecting H₀ when it is actually true.

### Type II Error (False Negative)
- Failing to reject H₀ when it is actually false.

---


```python
from scipy import stats
import pandas as pd
import numpy as np
```

```python
columns = [
    "Age", "Workclass", "Fnlwgt", "Education", "Education_Num",
    "Marital_Status", "Occupation", "Relationship", "Race", "Sex",
    "Capital_Gain", "Capital_Loss", "Hours_per_Week", "Native_Country", "Income"
]

df = pd.read_csv(
    "adult.data",
    header=None,
    names=columns,
    na_values="?",
    skipinitialspace=True
)
```

## 🔹 Hypothesis 1: Education vs Income (Chi-Square Test)

### Hypothesis 1 — Education vs Income

- **H₀:** Income is independent of education level.
- **H₁:** Income depends on education level.

**Test Used:** Chi-Square Test of Independence  
**Reason:** Both variables are categorical.


```python
edu_income_table = pd.crosstab(df["Education"], df["Income"])

chi2, p, dof, expected = stats.chi2_contingency(edu_income_table)

chi2, p
```

### Interpretation

- If **p-value ≤ 0.05**:
  - Reject the null hypothesis (H₀).
  - This indicates a **statistically significant association** between education level and income category.
  - In practical terms, **income distribution varies across different education levels**, suggesting that education is an important factor influencing whether an individual earns more than $50K.

- If **p-value > 0.05**:
  - Fail to reject the null hypothesis (H₀).
  - This suggests that **education level and income category are statistically independent**.
  - Any observed differences in income across education levels are likely due to random variation rather than a meaningful relationship.

## 🔹 Hypothesis 2: Hours per Week vs Income (t-Test)

### Hypothesis 2 — Hours per Week vs Income

- **H₀:** Mean hours worked per week is the same for both income groups.
- **H₁:** Mean hours worked per week differs between income groups.

**Test Used:** Independent t-test  
**Reason:** Numerical vs binary categorical variable.


```python
hours_low = df[df["Income"] == "<=50K"]["Hours_per_Week"].dropna()
hours_high = df[df["Income"] == ">50K"]["Hours_per_Week"].dropna()

t_stat, p_value = stats.ttest_ind(hours_low, hours_high, equal_var=False)

t_stat, p_value
```

### Interpretation

- If **p-value ≤ 0.05**:
  - Reject the null hypothesis (H₀).
  - This indicates a **statistically significant difference** in the average number of hours worked per week between individuals earning `<=50K` and those earning `>50K`.
  - In practical terms, **working hours per week are associated with income level**.

- If **p-value > 0.05**:
  - Fail to reject the null hypothesis (H₀).
  - This suggests there is **no statistically significant difference** in average weekly working hours between the two income groups.
  - Any observed difference may be due to random chance.

## 🔹  Hypothesis 3: Sex vs Income (Chi-Square Test)

### Hypothesis 3 — Sex vs Income

- **H₀:** Income is independent of sex.
- **H₁:** Income depends on sex.

**Test Used:** Chi-Square Test of Independence


```python
sex_income_table = pd.crosstab(df["Sex"], df["Income"])

chi2_sex, p_sex, dof_sex, exp_sex = stats.chi2_contingency(sex_income_table)

chi2_sex, p_sex
```

### Interpretation

- If **p-value ≤ 0.05**:
  - Reject the null hypothesis (H₀).
  - This indicates a **statistically significant association** between sex and income category.
  - Income distribution differs across gender groups.

- If **p-value > 0.05**:
  - Fail to reject the null hypothesis (H₀).
  - This suggests that **income category is independent of sex**, and any observed differences are not statistically significant.
